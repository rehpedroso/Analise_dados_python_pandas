#**Trabalhando com Planilhas do Excel**

```
#Importando a biblioteca
import pandas as pd
```

```
#Leitura dos arquivos
df1 = pd.read_excel("Aracaju.xlsx")
df2 = pd.read_excel("Fortaleza.xlsx")
df3 = pd.read_excel("Natal.xlsx")
df4 = pd.read_excel("Recife.xlsx")
df5 = pd.read_excel("Salvador.xlsx")
```

```
df5.head()
```

```
#juntando todos os arquivos
#concat - concatenar
df = pd.concat([df1,df2,df3,df4,df5])
```

```
#Exibindo as 5 primeiras linhas
df.head()
```

```
#Exibindo as 5 últimas linhas
df.tail()
```

```
#Amostra de 5 linhas do conjunto de dados de forma aleatoria
df.sample(5)
```

```
#Verificando o tipo de dado de cada coluna
df.dtypes
```

```
#Alterando o tipo de dado da coluna LojaID
df["LojaID"] = df["LojaID"].astype("object")
```

```
df.dtypes
```

```
df.head()
```

**Tratando valores faltantes**

```
#Consultando linhas com valores faltantes, valores null
df.isnull().sum()
```

```
#Substituindo os valores nulos pela média, em memória
df["Vendas"].fillna(df["Vendas"].mean(), inplace=True)
```

```
df["Vendas"].mean()
```

```
df.isnull().sum()
```

```
df.sample(15)
```

```
#Substituindo os valores nulos por zero em memória
df["Vendas"].fillna(0, inplace=True)
```

```
#Apagando as linhas com valores nulos em memória
df.dropna(inplace=True)
```

```
#Apagando as linhas com valores nulos com base apenas em 1 coluna
df.dropna(subset=["Vendas"], inplace=True)
```

```
#Removendo linhas que estejam com valores faltantes em todas as colunas
df.dropna(how="all", inplace=True)
```

**Criando colunas novas**

```
#Criando a coluna de receita
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```
df.head()
```

```
df["Receita/Vendas"] = df["Receita"] / df["Vendas"]
```

```
df.head()
```

```
#Retornando a maior receita
df["Receita"].max()
```

```
#Retornando a menor receita
df["Receita"].min()
```

```
#nlargest - retorna as 3 lojas com maior receita
df.nlargest(3, "Receita")
```

```
#nsamllest - retorna as 3 piores receitas do conjunto de dados
df.nsmallest(3, "Receita")
```

```
#Agrupamento por cidade e retornando a soma por cidade
df.groupby("Cidade")["Receita"].sum()
```

```
#Ordenando o conjunto de dados do maior para o menor, com base na receita
df.sort_values("Receita", ascending=False).head(10)
```

#**Trabalhando com datas**

```
#Trasnformando a coluna de data em tipo inteiro
df["Data"] = df["Data"].astype("int64")
```

```
#Verificando o tipo de dado de cada coluna
df.dtypes
```

```
#Transformando coluna de data em datetime
df["Data"] = pd.to_datetime(df["Data"])
```

```
df.dtypes
```

```
#Agrupamento por ano, analisando o desenvolvimento da receita a partir do ano
#df["Data"].dt.year - quer dizer que ele irá extrair apenas o ano da coluna 'data'
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```
#Criando uma nova coluna com o ano
df["Ano_Venda"] = df["Data"].dt.year
```

```
df.sample(5)
```

```
#Extraindo o mês e o dia
df["mes_venda"], df["dia_venda"] = (df["Data"].dt.month, df["Data"].dt.day)
```

```
df.sample(5)
```

```
#Retornando a data mais antiga
df["Data"].min()
```

```
#Calculando a diferença de dias
df["diferenca_dias"] = df["Data"] - df["Data"].min()
```

```
df.sample(5)
```

```
#Criando a coluna de trimestre
df["trimestre_venda"] = df["Data"].dt.quarter
```

```
df.sample(5)
```

```
#Filtrando as vendas de 2019 do mês de março
vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
```

```
vendas_marco_19.sample(20)
```

#**Visualização de dados**

```
#value_counts - contagem de linhas que determinada informacao aparcee no conjunto, ou seja, neste caso ele verifica
#quantas vezes determinada lojaId aparece no conjunto de dados
#ascending=False - traz as informacoes do maior para o menor
df["LojaID"].value_counts(ascending=False)
```

```
#Gráfico de barras
#para não aparecer essa linha em cima do grafico '<Axes: >', basta colocar um ';'
#df["LojaID"].value_counts(ascending=False).plot.bar();
df["LojaID"].value_counts(ascending=False).plot.bar()
```

```
#Gráfico de barras horizontais, não ordenados
df["LojaID"].value_counts().plot.barh();
```

```
#Gráfico de barras horizontais, hordenados
df["LojaID"].value_counts(ascending=True).plot.barh();
```

```
#Gráfico de Pizza
#A analise esta sendo feita atraves da receita por ano, com base nos dados do conjunto
df.groupby(df["Data"].dt.year)["Receita"].sum().plot.pie();
```

```
#Total vendas por cidade
df["Cidade"].value_counts()
```

```
#Adicionando um título e alterando o nome dos eixos
import matplotlib.pyplot as plt
df["Cidade"].value_counts().plot.bar(title="Total vendas por Cidade") #Titulo para o grafico, passo por parametro
plt.xlabel("Cidades") #Nome do eixo X
plt.ylabel("Total Vendas"); #Nome do eixo y
```

```
#Alterando a cor
df["Cidade"].value_counts().plot.bar(title="Total vendas por Cidade", color="purple") #Cor desejada, passando por parametro
plt.xlabel("Cidade")
plt.ylabel("Total Vendas");
```

```
#Alterando o estilo
plt.style.use("ggplot")
```

```
df.groupby(df["mes_venda"])["Qtde"].sum().plot(title = "Total Produtos vendidos vs mês")
plt.xlabel("Mês")
plt.ylabel("Total Produtos Vendidos")
plt.legend();
```

```
#Traz a quantidade de vendas realizadas por mes, porem juntando as informacoes dos anos que ha no ocnjunto de dados
df.groupby(df["mes_venda"])["Qtde"].sum()
```

```
#Selecionando apenas as vendas de 2019
df_2019 = df[df["Ano_Venda"] == 2019]
```

```
df_2019.groupby(df_2019["mes_venda"])["Qtde"].sum()
```

```
#Total produtos vendidos por mês
df_2019.groupby(df_2019["mes_venda"])["Qtde"].sum().plot(marker = "o") #marcador sendo passado por parametros,
#para identicar melhor as informacoes
plt.xlabel("Mês")
plt.ylabel("Total Produtos Vendidos")
plt.legend();
```

```
#Hisograma
plt.hist(df["Qtde"], color="orangered");
```

```
#Grafico de dispersão
plt.scatter(x=df_2019["dia_venda"], y = df_2019["Receita"]);
```

```
#Salvando os graficos em png
df_2019.groupby(df_2019["mes_venda"])["Qtde"].sum().plot(marker = "v")
plt.title("Quantidade de produtos vendidos x mês")
plt.xlabel("Mês")
plt.ylabel("Total Produtos Vendidos");
plt.legend()
plt.savefig("grafico QTDE x MES.png") #savefig - salve esta imagem, passando o nome desejado
#A imagem será salva dentro da pasta de arquivos (pasta lateral no google colab)
```
